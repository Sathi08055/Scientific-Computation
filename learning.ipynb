```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

num_points=20
a=-10
b=10
def my_function_vector(x,y):
    vx=y**2
    vy=x**2
    return vx, vy

def my_curl(f,x,y,h=1e-5):
    fyx=(f(x+h,y)[1]-f(x-h,y)[1])/(2*h)
    fxy=(f(x,y+h)[0]-f(x,y-h)[0])/(2*h)
    cz=fyx-fxy
    return cz/10

fig,ax=plt.subplots()
x=np.linspace(a,b,num_points)
y=np.linspace(a,b,num_points)
z=np.linspace(a,b,num_points)
x,y =np.meshgrid(x,y)
vx,vy=my_function_vector(x,y)
ax.quiver(x,y,vx,vy)
plt.title("Vector $-y \\hat i + x \\hat j$")
plt.show()
x=np.linspace(a,b,5)
y=np.linspace(a,b,5)
z=np.array([1,10])
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
x_mesh,y_mesh,z_mesh =np.meshgrid(x,y,z)
ax.quiver(x_mesh,y_mesh,z_mesh,0,0,my_curl(my_function_vector,x_mesh,y_mesh))
ax.view_init(elev=0)
plt.show()
```

```python
def my_function_vector(x,y):
    vx=x
    vy=y
    return vx, vy

def my_curl(x,y):
    vx=x/x
    vy=y/y
    return vx, vy

fig,ax=plt.subplots()
x=np.linspace(-50,50,15)
y=np.linspace(-50,50,15)
x,y =np.meshgrid(x,y)
vx,vy=my_function_vector(x,y)
ax.quiver(x,y,vy,vx)
plt.title("Vector $-y \\hat i + x \\hat j$")
plt.show()
```
